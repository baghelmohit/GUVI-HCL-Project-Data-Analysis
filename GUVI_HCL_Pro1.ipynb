```python
!pip install plotly pandas
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("/content/customer_support_tickets.csv")
df.head()
```

```python
# Convert time columns to datetime
df['First Response Time'] = pd.to_datetime(df['First Response Time'])
df['Time to Resolution'] = pd.to_datetime(df['Time to Resolution'])

# Filter only closed tickets (optional)
df = df[df['Ticket Status'] == 'Closed'].copy()

# Calculate resolution time in hours
df['Resolution Time (hrs)'] = (df['Time to Resolution'] - df['First Response Time']).dt.total_seconds() / 3600
```

```python
ticket_count = df.groupby(['Ticket Type', 'Ticket Channel']).size().reset_index(name='Ticket Count')

fig = px.bar(ticket_count,
               x='Ticket Type', y='Ticket Count',
               color='Ticket Channel',
               barmode='group',
               title='Ticket Count by Category and Channel')

fig.show()
```

```python
avg_resolution = df.groupby(['Ticket Type', 'Ticket Channel'])['Resolution Time (hrs)'].mean().reset_index()

fig2 = px.bar(avg_resolution,
              x='Ticket Type', y='Resolution Time (hrs)',
              color='Ticket Channel',
              barmode='group',
              title='Average Resolution Time by Ticket Type and Channel')

fig2.show()
```
